### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user" , sep='|' , index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(['occupation']).mean()['age']
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users[users['gender']=='M'].groupby(['occupation']).count()['gender'].sort_values(ascending =True)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
pd.DataFrame(data = {
                     'max':users.groupby(['occupation']).max()['age'],
                     'min':users.groupby(['occupation']).min()['age']
                    })
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).mean()['age']
```

### Step 8.  For each occupation present the percentage of women and men

```python
pd.DataFrame(data = {
    "Female Percetage":users[users['gender']=='F'].groupby(['occupation']).count()['gender'],
    "Male Percentage":users[users['gender']=='M'].groupby(['occupation']).count()['gender']
})
```
